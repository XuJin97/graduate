```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
np.random.seed(2020)
random.seed(2020)
N_female = 4000
N_male = 1000
N = 5000
y_female = np.random.normal(loc = 95.0, scale = 2, size = N_female)
y_male = np.random.normal(loc = 130.0, scale = 2, size = N_male)
y = np.hstack((y_female, y_male)) #combine two population
mu = np.mean(y)
```

```python
def one_sample_srs(n):
    sample_index = np.random.choice(N, n, replace = False)
    mean_i = np.mean(y[sample_index]) #sample mean
    return mean_i
for n in [200, 2000]:
    result_srs = np.zeros([1000, 1])
    for i in range(1000):
        result_srs[i, :] = one_sample_srs(n)
    plt.plot()
    plt.hist(result_srs)
    plt.axvline(mu, color = 'r') #abline of mu
    plt.savefig('sample_srs_'+str(n)+'.png')
    plt.show()
    plt.close()
```

```python
def one_sample_srs_seperate(n):
    sample_index_female = np.random.choice(N_female, n, replace = False)
    sample_index_male = np.random.choice(N_male, n, replace = False)
    y_sample = np.hstack((y_female[sample_index_female], y_male[sample_index_male])) #sampling seperately and combine together
    mean_i = np.mean(y_sample)
    return mean_i
for n in [100, 300]:
    result_srs_seperate = np.zeros([1000, 1])
    for i in range(1000):
        result_srs_seperate[i, :] = one_sample_srs_seperate(n)
    plt.plot()
    plt.hist(result_srs_seperate)
    plt.savefig('sample_srs_seperate_'+str(n)+'.png')
    plt.show()
    plt.close()
for n in [100, 300]:
    result_srs_seperate = np.zeros([1000, 1])
    for i in range(1000):
        result_srs_seperate[i, :] = one_sample_srs_seperate(n)
    plt.plot()
    plt.hist(result_srs_seperate)
    plt.axvline(mu, color = 'r')
    plt.savefig('sample_srs_seperate_line_'+str(n)+'.png')
    plt.show()
    plt.close()
```

```python
def one_sample_srs_seperate_prob(n):
    sample_index_female = np.random.choice(N_female, n, replace = False)
    sample_index_male = np.random.choice(N_male, n, replace = False)
    
    #weighted average of mean by gender probibility
    mu_hat_i = N_female / N * np.mean(y_female[sample_index_female]) + N_male / N * np.mean(y_male[sample_index_male]) 
    return mu_hat_i
for n in [100, 300]:
    result_srs_seperate_prob = np.zeros([1000, 1])
    for i in range(1000):
        result_srs_seperate_prob[i, :] = one_sample_srs_seperate_prob(n)
    plt.plot()
    plt.hist(result_srs_seperate_prob)
    plt.axvline(mu, color = 'r')
    plt.savefig('sample_srs_seperate_prob_'+str(n)+'.png')
    plt.show()
    plt.close()
```
